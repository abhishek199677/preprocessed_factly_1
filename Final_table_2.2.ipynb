```python
import pandas as pd

#reading csv and converting it into pandas dataframe
df = pd.read_csv(r"C:\Users\putta\Downloads\table-2.2.csv")
```

```python

 #Specify columns and rows to drop
df.drop([0,3,4,34,42,43,44,45,46,47,48,49,50], inplace = True)

#Drop rows with any NaN values

df.dropna(axis=0, how='all', inplace=True)
df.dropna(axis=1, how='all', inplace=True)


df
```

```python
#making header using indexing

df.columns = df.iloc[0]
df.head()

```

```python


# indexing to drop columns with all NaN values and a particular row
df = df.drop(df.columns[df.isna().all()], axis=1).drop([2])

# Display the updated DataFrame
print(df)

```

```python
df.head()
```

```python


# mapping dictionary for column renaming
column_mapping = {
    df.columns[0]: 'state',
    df.columns[-1]: 2011,
    df.columns[-2]: 2001,
    df.columns[-3]: 1991
}

# Rename columns using the mapping dictionary
df = df.rename(columns=column_mapping)

# Display the updated DataFrame
print(df)
```

```python

# Melting the DataFrame
melted = pd.melt(df, id_vars=['state'], var_name='year', value_name='population')

# Display the melted DataFrame
print(melted)
```

```python
#to add header columns
melted.insert(3,'unit','Population in Absolute Number')
melted.insert(4,'note','')

melted

```

```python

# Define a dictionary for replacements
replace_dict = {
    '&': 'and',
    'INDIA': 'All India',
    'D.and N.Haveli': 'Dadra and Nagar Haveli',
    'A.and N.Islands': 'Andaman and Nicobar Islands',
    'Chhatisgarh': 'Chhattisgarh',
}

# Perform replacements using the replace method
melted['state'] = melted['state'].replace(replace_dict, regex=True)

# Display the updated DataFrame
print(melted)

# Define a dictionary for special symbols replacements
symbol_replace_dict = {
    r'[*+]+': '',
}

# Perform replacements using the replace method with regex=True
melted['state'] = melted['state'].replace(symbol_replace_dict, regex=True)

# Display the updated DataFrame
print(melted)
```

```python
#cleaning data includes removing whitespaces, characters
# Applying str.strip() to all string columns in the DataFrame
melted = melted.applymap(lambda x: x.strip() if isinstance(x, str) else x)

display(melted)



```
